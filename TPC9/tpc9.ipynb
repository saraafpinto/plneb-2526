### 1º Modelo
vector_size=100, window=5, min_count=2, sg=0, epochs=5, workers=3

```python
from gensim.models import KeyedVectors
from gensim.models import Word2Vec

model_wv=KeyedVectors.load("word2vec.wordvectors")
```

```python
vector = model_wv['harry']
print(vector)
```

```python
'hermione' in model_wv
```

```python
!python -m gensim.scripts.word2vec2tensor -i model_harry.txt -o model_harry
```

```python
model_wv.most_similar('harry')
```

```python
model_wv.most_similar('rony')
```

```python
print(model_wv.similarity('harry','hermione'))
print(model_wv.similarity('harry','rony'))
print(model_wv.similarity('harry','voldemort'))
```

```python
result = model_wv.most_similar(positive=['draco', 'potter'], negative=['harry'])
print(result)
```

```python
result = model_wv.most_similar(positive=['draco', 'grifinória'], negative=['harry'])
print(result)
```
