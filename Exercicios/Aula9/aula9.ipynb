```python
from gensim.models import Word2Vec

# define training data
sentences = [['this', 'is', 'the', 'first', 'sentence', 'for', 'word2vec'],
            ['this', 'is', 'the', 'second', 'sentence'],
            ['yet', 'another', 'sentence'],
            ['one', 'more', 'sentence'],
            ['and', 'the', 'final', 'sentence'] ]


model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, sg=0, epochs=5, workers=3)

#o melhor valor pra vector_size é entre 100 e 300 
#window - vê as 5 palavras pra cada lado pra aprender a vizinhanca
#min_count=2 - se aparecer menos de 2 vezes nao vale a pena considerar aquela palavra
#sg - arquitetura
```

```python
model.wv['sentence']
```

```python
model.wv['Advanced']
#se procurar uma palavra que nao exista dda erro
```

```python
#podemos adicionar mais frases ao treino

more_sentences = [['Advanced', 'users', 'can', 'load', 'a', 'model', 'and', 'continue', 'training', 'it', 'with', 'more', 'sentences', 'ds', 'Advanced']]

model.build_vocab(more_sentences, update=True)
model.train(more_sentences, total_examples=model.corpus_count, epochs=model.epochs)
model.wv['Advanced']
```

```python
def analogy(x2,y1,x1):
    result = model_wv.most_similar(positive=[y1, x2], negative=[x1])
    return result[0][0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def display_pca_scatterplot(model, words=None):
    word_vectors = np.array([model[w] for w in words])
    twodim = PCA().fit_transform(word_vectors)[:,:2]
    plt. figure(figsize=(6,6))
    plt.scatter(twodim[:,0], twodim[:,1], edgecolors='k',c='r')
    for word,(x,y) in zip(words, twodim):
        plt.text(x+0.05, y+0.05, word)
```
